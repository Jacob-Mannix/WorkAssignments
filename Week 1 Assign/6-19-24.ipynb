```python
import pandas as pd
file_path = "../Data/imdb_top_1000.csv"
df = pd.read_csv(file_path)
df.info()
```

```python
# Get rid of useless columns
# Runtime, Poster_Link, Certificate, Meta_Score

def readincsv():    # made it a function for ease of use for the excersises
    try:
        # Define the file path
        file_path = "../Data/imdb_top_1000.csv"
        
        # Load the dataset
        df = pd.read_csv(file_path)
        
        # Drop the specified columns
        df.drop(columns=['Poster_Link', 'Runtime', 'Certificate', 'Meta_score'], inplace=True)
        
        return df
    except FileNotFoundError:
        print(f"Error: The file {file_path} was not found.")
        return None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

df = readincsv()

df.head()
```

```python
# Oldest movie

# Converting Released_Year from type object to type float
df = readincsv()

df['Released_Year'] = pd.to_numeric(df['Released_Year'], errors = 'coerce')

# dropping any rows with NAN values
df = df.dropna(subset=['Released_Year'])

# converting released year from float to int
df['Released_Year'] = df['Released_Year'].astype(int)

# sorting for the oldest movie 
oldest_movie = df[df['Released_Year']==df['Released_Year'].min()]

# printing it
print("oldest_movie: ")
print(oldest_movie[['Series_Title', 'Released_Year']])
```

```python
# Newest movie

# sorting for the oldest movie 
newest_movie = df[df['Released_Year']==df['Released_Year'].max()]

# printing it
print("newest_movie: ")
print(newest_movie[['Series_Title', 'Released_Year']])
```

```python
# Top 10 movies by IMDB rating
df = readincsv()

df_sorted = df.sort_values(by='IMDB_Rating', ascending=False)

top_10_movies = df_sorted.head(10)

print("Top 10 Movies by IMDB Rating: ")
print(top_10_movies[['Series_Title', 'IMDB_Rating']])
```

```python
# Top movie for each genre
df = readincsv()

# Ensure 'IMDB_Rating' is numeric, handling non-numeric values
df['IMDB_Rating'] = pd.to_numeric(df['IMDB_Rating'], errors='coerce')

# Split genres into separate rows
df['Genre'] = df['Genre'].str.split(', ')
df_exploded = df.explode('Genre')

# Remove duplicate genres
df_exploded = df_exploded.sort_values('IMDB_Rating', ascending=False).drop_duplicates('Genre')

# Display the top movie for each genre
print("Top Movie for Each Genre:")
print(df_exploded[['Genre', 'Series_Title', 'IMDB_Rating']])
```

```python
# Director with the most movies

df = readincsv()

# ensuring the director is of type string
df['Director'] = df['Director'].astype(str)

# counting number of movies for each director
director_counts = df['Director'].value_counts() # .value_counts() only counts unique instances, in this case unique director

# findind director with most movies
top_director = director_counts.idxmax() # idkmax() returns the index (in this case directors name) with the highest value (value being movies count)
top_director_count = director_counts.max()  # returns the maximum value (in this case it is the highest movie count)

print(f"Director with the most movies: {top_director}")
print(f"Number of movies: {top_director_count}")
```

```python
# Star with the most movies 

df = readincsv()

# Combine all star columns into a single column
star_columns = ['Star1', 'Star2', 'Star3', 'Star4']
df_stars = df[star_columns].melt(value_name='Star').drop('variable', axis=1)

# Drop NaN values from the combined star column
df_stars = df_stars.dropna().reset_index(drop=True)

# Count the number of movies for each star
star_counts = df_stars['Star'].value_counts()

# Identify the star with the most movies
top_star = star_counts.idxmax()
top_star_count = star_counts.max()

# Display the star with the most movies
print(f"Star with the most movies: {top_star}")
print(f"Number of movies: {top_star_count}")
```

```python
# Highest grossing movie for each genre 

df = readincsv()

# Ensure 'Gross' is numeric, handling non-numeric values
df['Gross'] = pd.to_numeric(df['Gross'].str.replace('[^\d.]', '', regex=True), errors='coerce')

# then drop any brocken cells aka NAN 
df = df.dropna(subset=['Gross'])

# Split Genres into seperate rows
df['Genre'] = df['Genre'].str.split(', ')
df_exploded = df.explode('Genre')

# group by genre and get highest grossing movie per genre
idx = df_exploded.groupby('Genre')['Gross'].idxmax()
highest_grossing_per_genre = df_exploded.loc[idx]

# drop the duplicate genres
highest_grossing_per_genre = highest_grossing_per_genre.drop_duplicates(subset=['Genre'])

# Display the highest-grossing movie for each genre
print("Highest Grossing Movie for Each Genre:")
print(highest_grossing_per_genre[['Genre', 'Series_Title', 'Gross']])
```

```python
# Lowest grossing movie for each director 

df = readincsv()

# Ensure 'Gross' is numeric, handling non-numeric values
df['Gross'] = pd.to_numeric(df['Gross'].str.replace('[^\d.]', '', regex=True), errors='coerce')

# then drop any brocken cells aka NAN 
df = df.dropna(subset=['Gross'])

# group by genre and get highest grossing movie per genre
idx = df_exploded.groupby('Director')['Gross'].idxmin()
lowest_gross_per_dir = df_exploded.loc[idx]

# drop the duplicate genres
lowest_gross_per_dir = lowest_gross_per_dir.drop_duplicates(subset=['Director'])

# Display the highest-grossing movie for each genre
print("Lowest Grossing Movie for Each Director:")
print(lowest_gross_per_dir[['Director', 'Series_Title', 'Gross']])
```

```python
# Save the dataframe as a parquet file 

# using the pyarrow install which has a to_parquet() command to save to that file type

file_path_parquet = 'imdb_top_1000.parquet'
df.to_parquet(file_path_parquet)
```

```python
# Aggregate the gross revenue for all 1000 movies.

df = readincsv()

# Ensure 'Gross' is numeric, handling non-numeric values
df['Gross'] = pd.to_numeric(df['Gross'].str.replace('[^\d.]', '', regex=True), errors='coerce')

# Drop rows where 'Gross' could not be converted
df = df.dropna(subset=['Gross'])

# Aggregate the gross revenue for all 1000 movies
total_gross_revenue = df['Gross'].sum()

# Display the total gross revenue
print(f"Total gross revenue for all 1000 movies: ${total_gross_revenue:,.2f}")
```

# Challenge - DICOM Image Processor 

```python
# Load a DICOM image into Python 
# Convert them into a gif 

# example from online as i coudnt figure it out myself in the end - not my own work just to help learn/study
# -----------------------------------------------------------------------------------------------------

import pydicom
from PIL import Image
import numpy as np
import os

# Define the folder path containing DICOM files and the output path for the GIF
dicom_folder_path = '../Data/ID00419637202311204720264'
output_gif_path = 'output_image.gif'

# Initialize a list to hold the frames
frames = []

# Process each DICOM file in the folder
for filename in sorted(os.listdir(dicom_folder_path), key=lambda x: int(os.path.splitext(x)[0])):
    if filename.endswith(".dcm"):
        file_path = os.path.join(dicom_folder_path, filename)
        
        # Load the DICOM image
        dicom = pydicom.dcmread(file_path)
        image_array = dicom.pixel_array

        # Normalize the image array to the range 0-255
        image_array = (255.0 / image_array.max() * (image_array - image_array.min())).astype(np.uint8)

        # Convert the numpy array to a PIL image
        image = Image.fromarray(image_array)
        
        # Add the image to the list of frames
        frames.append(image)

# Save the frames as an animated GIF
if frames:
    frames[0].save(output_gif_path, save_all=True, append_images=frames[1:], loop=0, duration=500)
    print(f"GIF saved to {output_gif_path}")
else:
    print("No DICOM files found in the folder.")

# -----------------------------------------------------------------------------------------------------
```
